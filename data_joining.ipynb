# Unión de datos
La ejecución de este notebook coloca en un solo archivo todos los registros de posición obtenidos para cada unidad de Transmetro

```python
import os
import pandas as pd
import warnings
warnings.simplefilter("ignore", UserWarning)
```

```python
import os
import pandas as pd

root_folder = './data'
output_folder = './joined_data'
os.makedirs(output_folder, exist_ok=True)

for unit_folder in sorted(os.listdir(root_folder)):
    unit_path = os.path.join(root_folder, unit_folder)
    
    if not os.path.isdir(unit_path):
        continue

    output_file = os.path.join(output_folder, f"{unit_folder}_joined.csv")
    if os.path.exists(output_file):
        print(f"🟢 {output_file} ya existe, se saltará.")
        continue

    print(f"Procesando unidad: {unit_folder}")
    all_data = []
    file_count = 0
    row_count = 00
    
    for file_name in sorted(os.listdir(unit_path)):
        if file_name.endswith('.xlsx'):
            file_path = os.path.join(unit_path, file_name)
            try:
                df = pd.read_excel(file_path, skiprows=3)
                df.dropna(how='all', inplace=True)
                
                if not df.empty:
                    df['archivo_origen'] = file_name
                    all_data.append(df)
                    file_count += 1
                    row_count += len(df)
                else:
                    print(f"⚠️ {file_name} está vacío después de limpiar.")
            except Exception as e:
                print(f"❌ Error leyendo {file_name}: {e}")
    
    if all_data:
        df_unit = pd.concat(all_data, ignore_index=True)
        df_unit.to_csv(output_file, index=False)
        print(f"✅ {unit_folder}: {file_count} archivos, {row_count:,} filas -> {output_file}")
    else:
        print(f"⚠️ No se encontraron datos útiles para la unidad {unit_folder}.")
    
    print("-" * 70)
```
