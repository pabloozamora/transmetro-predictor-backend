```python
from pathlib import Path
import pandas as pd
```

```python
DATA_WITH_FEATURES_DIR = Path("D:/2025/UVG/Tesis/repos/backend/data_with_features")
```

```python
def process_unit(unit):

    viterbi_file = DATA_WITH_FEATURES_DIR / unit / f"{unit}_trips_with_viterbi.csv"
    viterbi_df = pd.read_csv(viterbi_file)
    
    # Aquí iría el código para revisar los índices y reportar inconsistencias
    # Por ejemplo:
    inconsistencies = 0
    for i in range(len(viterbi_df) - 1):
        if viterbi_df.loc[i, 'idx_end'] + 1 != viterbi_df.loc[i + 1, 'idx_start']:
            print(f"Inconsistencia encontrada en unidad {unit}: idx_end {viterbi_df.loc[i, 'idx_end']} != idx_start {viterbi_df.loc[i + 1, 'idx_start']}")
            inconsistencies += 1
    if inconsistencies > 0:
        return True
    return False
```

```python
# Por cada archivo u_trips_with_viterbi, checkear cuantas veces el idx_end de un registro no es igual al idx_start del siguiente registro.
# Encontrar todas las unidades (carpetas en data with features)
units_with_inconsistencies = []

if not DATA_WITH_FEATURES_DIR.exists():
    print(f"No existe el directorio {DATA_WITH_FEATURES_DIR}. Fin.")
units = [p.name for p in DATA_WITH_FEATURES_DIR.iterdir() if p.is_dir() and p.name != "maps"]
if not units:
    print(f"No se encontraron unidades en {DATA_WITH_FEATURES_DIR}. Fin.")
for unit in units:
    print(f"Procesando unidad {unit}...")
    inconsistencies = process_unit(unit)
    if inconsistencies:
        units_with_inconsistencies.append(unit)
    
```

```python
if units_with_inconsistencies:
    print("Unidades con inconsistencias encontradas:")
    print(len(units_with_inconsistencies))
```

```python
# Check for empty directories in DATA_WITH_FEATURES_DIR
units_found_in_dir = [p.name for p in DATA_WITH_FEATURES_DIR.iterdir() if p.is_dir() and p.name != "maps"]
print(f"Unidades encontradas en el directorio: {len(units_found_in_dir)}")

for u in units_found_in_dir:
    if u not in units_with_inconsistencies:
        print(f"La unidad {u} con inconsistencias no se encontró en el directorio.")
```
